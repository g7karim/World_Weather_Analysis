```python
#Dependencies
import pandas as pd
import requests
import gmaps
import numpy as np
import random
import requests
import time
import datetime
from citipy import citipy 

# Import the API key.
from config import weather_api_key
from config import g_key
```

```python
#Create a new set of 2,000 random latitudes and longitudes
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
#lat_lngs = zip(lats, lngs)
#lat_lngs
coordinates = zip(lats, lngs)
coordinates

#Add coordinates to a list
coordinates_lst = list(coordinates)
coordinates_lst
```

```python
#Create list for cities
cities = []

#Get the nearest city using the citipy module.
for coordinate in coordinates_lst:
    city = citipy.nearest_city(coordinate[0], coordinate[1])
    city_name = city.city_name
    country_name = city.country_code
    cities.append(city_name)
    
cities
```

```python
#Perform an API call with the OpenWeatherMap.
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?"+"q=atlanta&"+"units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
city_weather = requests.get(url).json()
city_weather
```

```python
#Retrieve the following information from the API call: Latitude and longitude, Maximum temperature, Percent humidity
#Percent cloudiness, Wind speed & Weather description (for example, clouds, fog, light rain, clear sky).
#create for i loop 
city_data=[]
# Run an API request for each of the cities.
try:
# Parse the JSON and retrieve data.
    city_weather = requests.get(url).json()
    # Parse out the needed data.
    city = city_weather['name']
    city_lat = city_weather['coord']['lat']
    city_lng = city_weather['coord']['lon']
    city_max_temp = city_weather['main']['temp_max']
    city_humidity = city_weather['main']['humidity']
    city_clouds = city_weather['clouds']['all']
    city_wind = city_weather['wind']['speed']
    city_weather_descrip = city_weather['weather'][0]['description']
    city_rain = 0
    if 'rain' in city_weather_descrip: 
        try:
            city_rain = city_weather['rain']['3h']
        except:
            print('No rainfall data. Setting to 0in')

    city_snow = 0
    if 'snow' in city_weather_descrip:
        try:
            city_snow = city_weather['snow']['3h']
        except:
            print('No snowfall data. Setting to 0in')

    city_country = city_weather['sys']['country']
    # Convert the date to ISO standard.
    #city_date = datetime(city_weather['dt']).strftime('%Y-%m-%d %H:%M:%S')
    # Append the city information into city_data list.
    city_data.append({'City': city,
                      'Lat': city_lat,
                      'Lng': city_lng,
                      'Max Temp': city_max_temp,
                      'Humidity': city_humidity,
                      'Cloudiness': city_clouds,
                      'Wind Speed': city_wind,
                      'Weather Description': city_weather_descrip,
                      'Rainfall in. (last 3 hrs.)': city_rain,
                      'Snowfall in. (last 3 hrs.)': city_snow,
                      'Country': city_country,
                      #'Date': city_date
                     })
# If an error is experienced, skip the city.
except:
    print('City not found. Skipping...')
    pass

# Indicate that Data Loading is complete.
print('-----------------------------')
print('Data Retrieval Complete      ')
print('-----------------------------')
city_data
```

```python
#Add the data to a new DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = '../Weather_Database/Weather_Database.csv'

# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label='City_ID')
```
